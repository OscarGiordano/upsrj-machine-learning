# ORGANIZACIÓN Y ACCESO A DATOS
Este bloque define una base de datos como una lista de listas que almacena nombres, edades y animales.
Se utilizan diccionarios para indicar la posición de filas y columnas, lo que permite acceder de forma directa a un dato específico, en este caso el animal asociado a una persona.


```
base_de_datos=[
    ["Oscar",21,   "gatos"],
    ["paula",21,   "tiburones"],
    ["Jonatan",24, "Hurones"],
    ["kevin",22,   "periquitos"],
    ["lucero", 21, "perros"],
    ["israel",21,  "lobos"],
    ["sergio",21,  "perros"],
    ["yahir",21,   "leones"],
    ["diego",21,   "pinguinos"]

]
columnas ={"Nombre":0, "Edad":1, "Animal":2}
filas={"oscar":0, "paula":1,"jonatan":2, "kevin":3,"lucero":4,"israel":5,"sergio":6,"yahir":7,"diego":8}
print(base_de_datos[filas["kevin"]][columnas["Animal"]])
```

# CREACIÓN Y CONSULTA DE DATAFRAME
Este bloque convierte la base de datos en un DataFrame de Pandas, asignando nombres a las columnas y un índice numérico.
Posteriormente, se realiza una consulta para obtener el valor de la columna "Animal" correspondiente a una persona específica.


```
dF = pd.DataFrame(base_de_datos, columns=["Nombre", "Edad", "Animal"], index=range(1,10))
print(dF)
print(dF.Animal[dF['Nombre'] == 'Kevin'])
```
